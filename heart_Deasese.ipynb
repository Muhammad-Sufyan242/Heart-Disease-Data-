```python
ax = sns.countplot(x='sex', data=df)

ax.set_xticklabels(['Female', 'Male'])

for container in ax.containers:
    ax.bar_label(container)

plt.title('Male vs Female Count')
plt.xlabel('Gender')
plt.ylabel('Count')

plt.show()
```

```python
ax = sns.countplot(x='sex', hue='target', data=df)

ax.set_xticklabels(['Female', 'Male'])

for container in ax.containers:
    ax.bar_label(container)

plt.title('Heart Disease by Gender')
plt.show()
```

```python


def plotting(var, num):
    plt.subplot(2, 3, num)
    sns.histplot(df[var], kde=True)
    plt.title(var)

plt.figure(figsize=(12, 8))

plotting('age', 1)
plotting('trestbps', 2)
plotting('chol', 3)
plotting('thalach', 4)
plotting('oldpeak', 5)

plt.tight_layout()
plt.show()
```

```python
df['trestbps'] = df['trestbps'].astype(float)

trestbps_mean = df.loc[df['trestbps'] < 180, 'trestbps'].mean()

df.loc[df['trestbps'] >= 180, 'trestbps'] = trestbps_mean

df['trestbps'] = df['trestbps'].round(2)

```

```python
df['chol'] = df['chol'].astype(float)

trestbps_mean = df.loc[df['chol'] < 400, 'chol'].mean()

df.loc[df['chol'] >= 400, 'chol'] = trestbps_mean

df['chol'] = df['chol'].round(2)
```
